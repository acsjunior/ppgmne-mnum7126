# Mapas - Folium

```python
import pandas as pd
import folium
```

```python
city = pd.read_csv("cities.csv")
```

```python
city.head()
```

```python
city.shape
```

```python
latO = city.loc[0].at["city_lat"]
lonO = city.loc[0].at["city_lon"]
nameO = city.loc[0].at["city"]

print(nameO)

m = folium.Map(location=[city.loc[0].at["city_lat"], city.loc[0].at["city_lon"]])

#plot Origin
folium.Marker(
    location=[latO, lonO],
    popup=nameO,
    icon=folium.Icon(color="blue"),
            ).add_to(m)

# Just to zoom the maps' bounds
m.fit_bounds(m.get_bounds())

#plot map
m
```

```python
latO = city.loc[0].at["city_lat"]
lonO = city.loc[0].at["city_lon"]
nameO = city.loc[0].at["city"]

m = folium.Map(location=[city.loc[0].at["city_lat"], city.loc[0].at["city_lon"]])

#plot city
for i in range(city.shape[0]):
    latO = city.loc[i].at["city_lat"]
    lonO = city.loc[i].at["city_lon"]
    nameO = city.loc[i].at["city"]

    folium.Marker(
        location=[latO, lonO],
        popup=nameO,
        icon=folium.Icon(color="blue"),
                ).add_to(m)

# Just to zoom the maps' bounds
m.fit_bounds(m.get_bounds())

#plot map
m
```

```python
cityRS = city[city["state"]=="RS"]
cityRS.reset_index(inplace=True, drop = True )
cityRS.head()
```

```python
latO = cityRS.loc[0].at["city_lat"]
lonO = cityRS.loc[0].at["city_lon"]
nameO = cityRS.loc[0].at["city"]

m = folium.Map(location=[cityRS.loc[0].at["city_lat"], cityRS.loc[0].at["city_lon"]])

#plot Origin
for i in range(cityRS.shape[0]):
    latO = cityRS.loc[i].at["city_lat"]
    lonO = cityRS.loc[i].at["city_lon"]
    nameO = cityRS.loc[i].at["city"]

    folium.Marker(
        location=[latO, lonO],
        popup=nameO,
        icon=folium.Icon(color="green"),
                ).add_to(m)

# Just to zoom the maps' bounds
m.fit_bounds(m.get_bounds())

#plot map
m
```

```python
latO = cityRS.loc[0].at["city_lat"]
lonO = cityRS.loc[0].at["city_lon"]
nameO = cityRS.loc[0].at["city"]

latRef = cityRS.loc[6].at["city_lat"]
lonRef = cityRS.loc[6].at["city_lon"]

m = folium.Map(location=[cityRS.loc[0].at["city_lat"], cityRS.loc[0].at["city_lon"]])

#plot Origin
for i in range(cityRS.shape[0]):
    latO = cityRS.loc[i].at["city_lat"]
    lonO = cityRS.loc[i].at["city_lon"]
    nameO = cityRS.loc[i].at["city"]

    folium.Marker(
        location=[latO, lonO],
        popup=nameO,
        icon=folium.Icon(color="green"),
                ).add_to(m)

    #plot lines
    coords = [[latO,lonO],[latRef,lonRef]]
    folium.PolyLine(coords, color="green").add_to(m) 

# Just to zoom the maps' bounds
m.fit_bounds(m.get_bounds())

#plot map
m
```

```python
m = folium.Map(location=[45.372, -121.6972], zoom_start=12, tiles="Stamen Terrain")


latO = cityRS.loc[0].at["city_lat"]
lonO = cityRS.loc[0].at["city_lon"]
nameO = cityRS.loc[0].at["city"]

latRef = cityRS.loc[6].at["city_lat"]
lonRef = cityRS.loc[6].at["city_lon"]

m = folium.Map(location=[cityRS.loc[0].at["city_lat"], cityRS.loc[0].at["city_lon"]], tiles="Stamen Terrain")

#plot Origin
for i in range(cityRS.shape[0]):
    latO = cityRS.loc[i].at["city_lat"]
    lonO = cityRS.loc[i].at["city_lon"]
    nameO = cityRS.loc[i].at["city"]

    folium.Marker(
        location=[latO, lonO],
        popup=nameO,
        icon=folium.Icon(color="green"),
                ).add_to(m)

    #plot lines
    coords = [[latO,lonO],[latRef,lonRef]]
    folium.PolyLine(coords, color="green").add_to(m) 

# Just to zoom the maps' bounds
m.fit_bounds(m.get_bounds())

#plot map
m
```

```python
url = ("https://raw.githubusercontent.com/python-visualization/folium/main/examples/data")

# fronteiras das cidades do brasil 
# https://github.com/tbrugz/geodata-br
# fronteiras dos estados brasileiros 
# https://gist.githubusercontent.com/ruliana/1ccaaab05ea113b0dff3b22be3b4d637/raw/196c0332d38cb935cfca227d28f7cecfa70b412e/br-states.json

state_geo = f"{url}/us-states.json"
state_unemployment = f"{url}/US_Unemployment_Oct2012.csv"
state_data = pd.read_csv(state_unemployment)

m = folium.Map(location=[48, -102], zoom_start=3)

folium.Choropleth(
    geo_data=state_geo,
    name="choropleth",
    data=state_data,
    columns=["State", "Unemployment"],
    key_on="feature.id",
    fill_color="YlGn",
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name="Unemployment Rate (%)",
).add_to(m)

folium.LayerControl().add_to(m)

m
```
