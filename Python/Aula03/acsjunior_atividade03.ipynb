```python
import pandas as pd

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('nba.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
# Exibindo a contagem de jogadores por faixa de altura em cada time:

pd.crosstab(df['Team'], df['Height'])
```

```python
# Exibindo o peso médio dos jogadores por altura em cada time:

df_pivot = pd.pivot_table(
    df, values='Weight', 
    index=['Team'], 
    columns=['Height'], 
    aggfunc='mean'
)

df_pivot
```

```python
# Exibindo o peso médio dos jogadores por altura em cada time / posição:

df_pivot = pd.pivot_table(
    df, values='Weight', 
    index=['Team', 'Position'], 
    columns=['Height'], 
    aggfunc='mean'
)

df_pivot
```
